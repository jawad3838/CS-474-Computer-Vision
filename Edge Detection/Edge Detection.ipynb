```python
# import libraries
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
import cv2
```

```python
#Task1
img = cv2.imread('lena.jpg',0)
blur = cv2.GaussianBlur(img,(5,5),0)
blur2 = cv2.GaussianBlur(img,(15,15),0)
edge1 = blur - blur2
edge2 = blur2 - blur

cv2.imwrite('dog1.jpg', edge1)
plt.imshow(edge1,cmap = 'gray')
plt.title('DOG 1')
plt.show()

cv2.imwrite('dog2.jpg', edge2)
plt.imshow(edge2,cmap = 'gray')
plt.title('DOG 2')
plt.show()
```

```python
#Task 2
img = cv2.imread('lena.jpg',0)
gaussian = cv2.GaussianBlur(img,(5,5),0)
log = cv2.Laplacian(gaussian,cv2.CV_64F)

sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)  
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)  

cv2.imwrite('log.jpg', log)
plt.imshow(log,cmap = 'gray')
plt.title('LOG')
plt.show()

cv2.imwrite('sobelX.jpg', sobelx)
plt.imshow(sobelx,cmap = 'gray')
plt.title('sobelX')
plt.show()

cv2.imwrite('sobelY.jpg', sobely)
plt.imshow(sobely,cmap = 'gray')
plt.title('sobelY')
plt.show()
```
