```python
import cv2 
import numpy as np 
from matplotlib import pyplot as plt
import random
import imutils
```

```python
image = cv2.imread('image.jpeg') 
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
grayinfloat32 = np.float32(gray)
#img,size of neighbourhood,Aperture parameter of Sobel derivative used,Harris detector free parameter in the equation.
harrisCorners = cv2.cornerHarris(grayinfloat32, 2, 3, 0.07) 
harrisCorners = cv2.dilate(harrisCorners, None) 
image[harrisCorners > 0.01 * harrisCorners.max()]=[0, 0, 255] 

cv2.imwrite('harris_corner.jpg', image)
harris_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(harris_img)
plt.title('Harris Corner')
plt.show()
```

```python
image = cv2.imread('image.jpeg') 
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
sift = cv2.xfeatures2d.SIFT_create()
keypoints, descriptors = sift.detectAndCompute(gray, None)
image = cv2.drawKeypoints(image, keypoints, None)

cv2.imwrite('sift_keypoints.jpg', image)
sift_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(sift_img)
plt.title('Sift Features')
plt.show()
```

```python
image = cv2.imread('image.jpeg') 
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
surf = cv2.xfeatures2d.SURF_create()
keypoints, descriptors = surf.detectAndCompute(gray, None)
image = cv2.drawKeypoints(image, keypoints, None)

cv2.imwrite('surf_keypoints.jpg', image)
surf_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(surf_img)
plt.title('Surf Features')
plt.show()
```

```python
# SIFT provides the most accurate feature detection, while SURF is faster but less accurate.
```
